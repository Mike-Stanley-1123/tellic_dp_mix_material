This notebook is purely meant to try reading some flow cytometry data files from the standard format `.fcs`. There is a package to do this called `FlowCytometryTools` that allegedly allows the reading of these files. Hoping that this package will allow me to read the data files in the `fcm` package to even start to reproduce the results from the Cron paper.

Use the link here: https://eyurtsev.github.io/FlowCytometryTools/tutorial.html

```python
import FlowCytometryTools
from FlowCytometryTools import FCMeasurement
import matplotlib.pyplot as plt; plt.style.use('fivethirtyeight')
```

```python
# specify path to data
datafile = '/Users/mikestanley/flow_cytometry/fcm-0.9.1/sample_data/3FITC_4PE_004.fcs'
```

```python
# read the datafile in
sample = FCMeasurement(ID='test sample', datafile=datafile)
```

```python
sample.data
```

### Plot a bit

```python
# put the data into more easily accessible frame
data = sample.data
```

```python
fig, ax = plt.subplots(4, 1, figsize=(8, 10))

# FSC-H
ax[0].hist(data['FSC-H'], bins=100, density=True)

# SSC-H
ax[1].hist(data['SSC-H'], bins=100, density=True)

# FL1-H
ax[2].hist(data['FL1-H'], bins=100, density=True)

# FL2-H
ax[3].hist(data['FL2-H'], bins=100, density=True)

plt.show()
```
